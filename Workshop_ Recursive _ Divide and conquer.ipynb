Although merge sort has a better Big-O than selection sort, selection sort can be faster for smaller inputs.

Rewrite `merge_sort(A, min_size)` such that sub-arrays smaller than an input parameter `min_size` are sorted with our `selection_sort` from the lecture `algorithms intro`.

Time the difference between pure merge sort and this new algorithm. Is it faster? Why or why not?

````
import time
import numpy as np 
def linear_search(arr):
  """
  Find the index of the minimum element
  AKA argsort
  """
  # initialize current best to +infinity
  # So any element beats it
  current_min = float('inf')
  current_min_idx = 0
  for i in range(len(arr)):
    if arr[i] < current_min:
      current_min = arr[i]
      current_min_idx = i
  return current_min_idx

def selection_sort(arr):
  """Selection sort"""
  n_sorted = 0
  while n_sorted < len(arr):
    # Get the index of the min of remaining elements
    # Since argsort returns based on array, we correct result
    # with `+ n_sorted`
    min_idx = linear_search(arr[n_sorted:]) + n_sorted
    # Swap minimum element with leftmost remaining element
    to_swap = arr[n_sorted]
    arr[n_sorted] = arr[min_idx]
    arr[min_idx] = to_swap
    # Increment and restart
    n_sorted += 1
  return arr


def merge(left, right):
  res = []
  # Zip in together left and right parts
  while len(left)>0 and len(right)>0: 
      if left[0]<right[0]: 
          res.append(left[0]) 
          left.pop(0)
      else: 
          res.append(right[0]) 
          right.pop(0)
  # Copy in remaining elements of left and right
  # (if there are any)
  for i in left: 
      res.append(i) 
  for i in right: 
      res.append(i)
  return res

def merge_sort(A): 
    size = len(A)
    if size > 1:
      m = size // 2
      left = merge_sort(A[m:]) 
      right = merge_sort(A[:m])
      return merge(left, right)
    else:
      return A

# Modified merge sort 
def merge_sort2(A,min_size): 
    size = len(A)
    if size>min_size :
        m = size // 2
        left = merge_sort2(A[m:],min_size) 
        right = merge_sort2(A[:m],min_size)
        return merge(left, right) 
    else:
      return selection_sort(A)
    

arr =[33, 1, 55, 2343, -232, 344, 2, 53, -4, 923]



k=20
L=np.random.random(100000)
L=list(L)


start = time.time()
a1=merge_sort(L)
end = time.time()
print('Regular merge_sort time:',end - start)


start = time.time()
a2=merge_sort2(L,k)

end = time.time()
print('Modified merge_sort time:',end - start)

````

# 3. Recursive sum

Write a function that uses recursion to compute the sum of an array or list of numbers

```
recursive_sum([2, 4, 5, 6, 7])

output: 24
```

````
L=[2, 4, 5, 6, 7]
def recursive_sum(L):
    Sum=0
    if len(L)==0:
        return Sum
    Sum+=L[0]
    L.pop(0)
    Sum+=recursive_sum(L)
    return Sum     

print(recursive_sum(L))
````

# 4. Recursive denominators

Write a Python program that uses recursion to find the greatest common divisor (gcd) of two integers.

```

output : 2
```

````
def recursive_gcd(x,y):
    #The euclidian algorithm 
    q=x//y
    r=x%y
    if r==0:
        return y
    return recursive_gcd(y,r)

    
# I used this site to test my function   
#http://www.alcula.com/calculators/math/gcd/#gsc.tab=0 

print(recursive_gcd(150_497_424,21_762_328))


print(recursive_gcd(12,14))
````

# 5. Recursive power function

Write a recursive function to calculate the value of 'a' to the power 'b'. 

```
recursive_pow(3, 4)

output: 81
```

````
def recursive_pow(a,n):
    if n==0:
        return 1
    else:
        return a*recursive_pow(a,n-1)
print(recursive_pow(3,4))
````
